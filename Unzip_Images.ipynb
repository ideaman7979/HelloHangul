```python
import tensorflow as tf
import os
```

```python
def unzip_image(file_path, dest_path):
    # read files 
    file_list=os.listdir(file_path)
    
    for file_no in range(len(file_list)):
        # file open
        myfile=file_list[file_no]
        f = open(file_path+ myfile, 'rb')
        
        # Read Header
        file_header = f.read(8) 
        
        # Read and create Image files.
        i=0;
        while True : 
        # read code & check EoF
            code = f.read(2)
            if (len(code) < 2) :
                break
            # Create New file
            new_file_name = myfile+"_"+str(i)   
            writer = open(dest_path+new_file_name, 'wb')
            #Write class id and code of image
            class_id = file_no.to_bytes(2,'big')
            writer.write(class_id)
            writer.write(code)
            
            width=f.read(1)
            writer.write(width)
            
            height = f.read(1)
            writer.write(height)
            
            reserved = f.read(2)
            
            image_bytes=f.read(width[0]*height[0])
            cnt = writer.write(image_bytes)
            writer.close()
        #class_id(2bytes)+code(2bytes)+width(1byte)+height(1byte)+image_bytes(64)
            i+=1
            
        f.close()
        print("unzip is done")
```
